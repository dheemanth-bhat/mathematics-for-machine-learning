## Summary

1. Binomial Distribution: **Models the number of successes** in a fixed number of trials.
2. Poisson distribution: **Models the number of events** in a fixed interval of time.
3. Exponential distribution: **Models the time** between consecutive events.
4. Geometric Distribution: **Models the number of trials** it takes to get the first success.
5. Normal distribution: **Models outcomes** that cluster symmetrically around a mean with a bell-shaped curve.
6. Uniform distribution: **Models outcomes** that are equally likely within a given range.

## 1 Binomial Distribution

### Type

Discrete Probability Distribution

### Formulas

Given $p$ the probability of success, Binomial distribution helps us in calculating $k$ successes from $n$ trials.

$$
\large
\begin{aligned}
P(X = k) = {}^nC_k \cdot p^k \cdot (1 - p)^{n - k}
\end{aligned}
$$

#### Expected Value

$$
\large
\begin{aligned}
E[X] = n \cdot p
\end{aligned}
$$

#### PMF and CDF

```python
import scipy as sp

p =   # Probability of success
n =   # Number of trails
x =   # value of Random variable X

sp.stats.binom.pmf(k=x, p=p, n=n)
sp.stats.binom.cdf(k=x, p=p, n=n)
```

#### Variance

$$
\large
\begin{aligned}
\sigma^2(x) = n \cdot p \cdot (1 - p)
\end{aligned}
$$

## 2 Bernoulli Trial

### Formulas

Given $p$ the probability of success, Bernoulli Trial helps us in calculating $k$ successes from $1$ trials.

$$
\large
\begin{aligned}
P(X = k) = p^k \cdot (1 - p)^{1 - k}
\end{aligned}
$$

#### Expected Value

$$
\large
\begin{aligned}
E[X] = p
\end{aligned}
$$

$$
\large

### Confidence Interval (CI)

#### Formula

Formula to calculate Confidence Interval:

$$
\large
\begin{aligned}
\text{Confidence Interval} = \text{Point Estimate} \pm \text{Margin of Error}
\end{aligned}
$$

### CI Using CLT

#### Formula

Formula to calculate Confidence Interval for population mean using Central Limit Theorem:

$$
\large
\begin{aligned}
\text{Confidence Interval} &= \biggl[\bar{X} \pm Z \cdot \frac{\sigma}{\sqrt{n}} \biggr] \\[10pt]
                           &= \biggl[\bar{X} - Z \cdot \frac{\sigma}{\sqrt{n}}\;\;,\quad\bar{X} + Z \cdot \frac{\sigma}{\sqrt{n}} \biggr]
\end{aligned}
$$

Where,

- $\bar{X}$ - Mean of the Sample means - is $\text{Point Estimate}$
- $Z \cdot \text{SE}$ - is $\text{Margin of Error}$

#### Approach #1

Two step process to compute Confidence Interval for population mean:

1. Find z-score for margin error.
2. Compute interval using margin of error.

### CI Using Bootstrapping

#### Steps to Calculate CI using Bootstrapping

- **STEP #1**: Draw **Bootstrap sample** from population dataset.
- **STEP #2**: Calculate required sample statistic (mean, median, SD, etc.) from Bootstrap sample.
