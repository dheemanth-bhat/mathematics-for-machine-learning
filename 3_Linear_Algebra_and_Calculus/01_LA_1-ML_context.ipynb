### 1.2 Key terms

#### Data / Dataset

1. Dataset contains **data-points** i.e., each row/observation in the dataset is a data-point.
2. ML Datasets have two components $X$ (multiple columns) and $Y$ (single column).
3. Datasets with two components $X$ and $Y$ are called as historical (or labeled) dataset.
4. Each data-point can be represented as a coordinate on a Cartesian coordinate system (for example Cartesian plane).

#### X vs Y

Other terms used for $X$ and $Y$ components:

```
X - Features     | Input Variables | Independent variables | Predictors
Y - Label/Target | Output Variable | Dependent variable    | Predicted variables
```

#### Features

1. Features are **Independent Variables**.
2. Each feature is a **dimension** on a Cartesian Plane.
3. Two dimensional data can be visualized using **scatter plot** on a Cartesian Plane.
4. Three dimensional data can be visualized on a Cartesian Space.
5. Techniques like **TSNE**, **PCA** can be used for visualizing higher dimensional (> 3D) data.

#### Label

1. Label is a **Dependent Variable**.
2. A Label identifies to which class a data-point belong to.
3. A data-point can have at most one label.
4. There can be two or more types of labels in a dataset.

#### Classifier

1. A Classifier segregates the data-points into two or more groups (depending on the number of classes).
2. A Classifier is also called as Separator or **Decision Boundary**.

There are different types of Separators. Two basic types of Separators are:

1. **Liner Separator**: A straight line separating data-points belonging to two classes.
2. **Non-liner Separator**: A complex separator for non-linear data.

#### Linear Separator

- A Classifier of type Linear Separator is a **straight line** that segregates the data-points into two or more groups.
- A straight line is represented in Coordinate Geometry using below equation:

$$
\large
y = m \cdot x \;+\; c
$$

##### Types of Classification

Based on the number of types of labels in the dataset, there are two types of classification problems:

1. **Binary Classification**: Two types of labels in the dataset.
2. **Multi-class classification**: Two or more types of labels in the dataset.

## 2 Training Pipeline

A Training pipeline (a.k.a Learning process) is a cyclic process that contains below three components resulting in an ML Model.

1. Mapping Function
2. Loss Function
3. Optimization

### 2.1 Mapping Function

#### Definition

**Mapping function** := A function $f$ that takes $x$ as an input gives $y$ as an output.

Mathematically a mapping function can be represented as:

$$
\large
y = f(x)
$$

#### Properties

1. In Mapping function we choose a geometric shape.
2. Internally a Mapping function simply uses a classifier, say a linear line, to map the input $x$ to a label $y$.
3. Initially $f(x)$ predicts values randomly hence we need a **loss function** to quantify the "cost" of wrong classification.
4. Concepts from Linear Algebra and Coordinate Geometry are used in constructing a Decision Boundary used inside mapping function.

### 2.2 Loss Function

#### Definition

**Loss function** := Function that measures how close the predicted value from $f(x)$ is to the actual label $y$.  

#### Properties

1. Purpose of Loss function (also called as Cost Function) is to tell us "How off are we from reality?". 
2. Loss function only measures how "wrong" (delta) the predicted value is from the actual value.
3. Loss function just outputs an error metric.
4. We need **optimization** to improvise the loss function.
5. Improvising a loss function is the process of minimizing the error.

### 2.3 Optimization

#### Definition

Optimization := Process of finding best parameters to improvise/minimize the loss function.

#### Properties

1. Optimizing a classifier is like finding a **best fit line** (among many possible lines) that segregates the data-points based on their class labels.
2. Here a classifier is an ML Model or a line or a mathematical equation.
3. Optimization helps us find the best parameters of an ML model resulting in a "best fit line".
4. Optimization helps to generalize the model for future unseen data (rather than exactly fitting a line on a dataset).
5. Concepts like maximization, minimization from Calculus are used in optimizing a Loss function.

#### What is an ML Model?

An ML model is a parameterized mathematical function that maps input features to predicted targets.

## 3 Coordinate Geometry

### 3.1 Cartesian Coordinate System

#### Cartesian Plane

1. The Cartesian coordinate system is a grid-based mathematical tool used in precisely locating points in a space using numerical coordinates.
2. In ML context, these points are the data-points in the Dataset.
3. Cartesian coordinate system in 2D is called as **Cartesian Plane**.
4. While Cartesian plane is bound to only 2D space, Cartesian coordinate system is not limited to 2 dimensions.
5. Cartesian coordinate system in 3D is called as **Cartesian Space**.
6. Cartesian coordinate system in $n$D is called as **Euclidean Space**.

#### Axis

1. A Cartesian plane is a two-dimensional grid formed by the intersection of two perpendicular lines called as Axes:
2. There are two types of Axes:
   1. Horizontal $\text{X-axis}$
   2. Vertical $\text{Y-axis}$
3. These lines intersect at the origin, denoted as $0$.
4. Axes are numbered lines to pinpoint the exact location of any point using ordered pairs $(\,x\,,\,y\,)$ called as **coordinates**.
5. In $(\,x\,,\,y\,)$:
   1. $x$ is called as x-coordinate
   2. $y$ is called as y-coordinate

#### Quadrants

The intersecting axes divide the Cartesian Plane into four sections known as **Quadrants** (I, II, III, and IV).

#### Dimensions

- Two dimensions (on Cartesian plane): X-axis and Y-axis
- Three dimensions (on Cartesian space): X-axis, Y-axis and Z-axis

#### Line

A Straight line requires at least two coordinates: $(x_1,\, y_1)$ and $(x_2,\, y_2)$ to be plotted on a Cartesian plane.

#### Slope-Intercept Form

Equation for a single straight line in Slope-Intercept form:

$$
\large
y = m \cdot x \;+\; c
$$

Where,

- $m$ - Slope (how steep the line is).
- $c$ - Intercept (y-coordinate where line intersects with y-axis).
- $x$ - Independent variable (input).
- $y$ - Dependent variable (output).

Slope-Intercept Form: https://www.desmos.com/calculator/9c9pyyysir

### 3.2 Slope

#### What is a Slope?

- In Slope-Intercept form $m$ represents the Slope.
- Slope $m$ is the measure of rate-of-change of $y$ for every one-unit increase in $x$.

![Slope Intercept](images/slope_intercept.png)

> **Note**:
> 
> In the above diagram, $y$ is representing Dependent variable, its the output of function $f(x) = m\cdot{x} \;+\; c$.

#### Properties

1. Slope tells "How **steep** the line is?".
2. Altering the Slope value rotates the line around y-axis with y-intercept as center of rotation.
3. Slope is determined using $\tan \uptheta$, where $\uptheta$ in the angle between line and x-axis.
4. Green Line in the above figure will be perfectly parallel to y-axis when slope $m = \infty$
5. Green Line in the above figure will be perfectly parallel to x-axis when slope $m = 0$

Oscillating line example: https://www.desmos.com/calculator/hoitbc8jnb

#### Formula

##### Vector Interpretation of Slope

Given two points: $(x_1,\,y_1)$ and $(x_2,\, y_2)$ on a line, formula to calculate slope of the line $m$:

$$
\large
m = \frac{y_2 - y_1}{x_2 - x_1} = \frac{\Delta\;y}{\Delta\;x}
$$

##### Trigonometric Interpretation of Slope

Trigonometric method to calculate slope when the angle of inclination $\uptheta$ of line is known:

$$
\large
m = \tan \uptheta
$$

Alternate definition of Slope can be defined as Change in y-axis when $x$ changes by 1 unit.

#### Correlation between x and y

- if both $\Delta x$ and $\Delta y$ are changing in the same direction then its positive correlation, hence **slope is positive**.
- if $\Delta x$ and $\Delta y$ are changing in the opposite direction then its negative correlation, hence **slope is negative**.
- if $\Delta x = 0$ then the line is vertical and slope is $\infty$ or **slope is undefined**.
- if $\Delta y = 0$ then the line is horizontal and **slope is zero**.

- Negative Slope: https://www.desmos.com/calculator/shdewjcjxv

### 3.3 Intercept

#### What is the Intercept?

- Intercept $c$ is the y-coordinate value where the line intersects with the y-axis.

#### Properties

1. In Slope-Intercept form $c$ represents the Intercept.
2. Altering the Intercept value shifts the line away/towards the origin $(0, 0)$.
3. When Intercept is zero line passes through the origin.

Shifting line example: https://www.desmos.com/calculator/9vy9rsb42f

#### Formula

Given two points: $(x_1,\,y_1)$ and $(x_2,\, y_2)$ on a line, formulas to calculate intercept of the line $c$:

$$
\large
\begin{aligned}
c &= y_1 - (m \cdot x_1) \\[5pt]
  &\quad\text{or} \\[5pt]
c &= y_2 - (m \cdot x_2)
\end{aligned}
$$

#### y-intercept

- When $x = 0$, Intercept $c$ is called as y-intercept.
- When $x = 0$, the line is parallel to x-axis and y-intercept is where it touches the y-axis.
- When $x = 0$, intercept $c$ be considered as the distance of straight line from the origin.
- y-intercept: https://www.desmos.com/calculator/jjssblyjnu.

$$
\large
\begin{aligned}
y &= m \cdot x \;+\; c \\[3pt]
y &= m \cdot 0 \;+\; c \\[3pt]
y &= c
\end{aligned}
$$

### 3.4 Rules of Lines

Rules of Lines are some of the principles of coordinate geometry.

#### Parallel Lines Rule

$$
\large
m_1 = m_2
$$

- When two lines are parallel to each other: $m_1 = m_2$ 
- Parallel lines: https://www.desmos.com/calculator/3maxlgmlb8

#### Perpendicular Lines Rule

$$
\large
m_1 \times m_2 = -1
$$

- When two lines are perpendicular to each other: $m_1 * m_2 = -1$
- Perpendicular lines have slopes that are negative reciprocals of each other.
- Negative reciprocal slopes: https://www.desmos.com/calculator/k7kzcmblhj

#### Quiz #1

Given the following equations of two lines, both of which are parallel to each other what is the slope of both lines? 

- Line 1: $3x - 2y + 6 = 0$
- Line 2: $9x - 6y - 18 = 0$

A. Slope = -3/2 for both lines.  
B. Slope = -2/3 for both lines.  
C. Slope = 3/2 for both lines.  
D. Slope = 2/3 for both lines.   

##### Solution

Calculate slope and intercept for Equation #1.

$$
\large
3x - 2y + 6 = 0 \\[5pt]
$$

$$
\large
\begin{aligned}
-2y &= -3x \;+\; -6 \\[5pt]
  y &= \frac{-3x \;+\; -6}{-2} \\[5pt]
  y &= \frac{\cancel{-}3}{\cancel{-}2}x \;+\; \frac{\cancel{-}6}{\cancel{-}2} \\[5pt]
  y &= \frac{3}{2}x \;+\; 3
\end{aligned}
$$

Calculate slope and intercept for Equation #2.

$$
\large
9x - 6y - 18 = 0 \\[5pt]
$$

$$
\large
\begin{aligned}
-6y &= -9x \;+\; -18 \\[5pt]
  y &= \frac{-9x \;+\; -18}{-6} \\[5pt]
  y &= \frac{\cancel{-}9}{\cancel{-}6}x \;+\; \frac{\cancel{-}18}{\cancel{-}6} \\[5pt]
  y &= \frac{3}{2}x \;+\; 3
\end{aligned}
$$

Option C. Slope = $\displaystyle \frac{3}{2}$ for both lines.

### 3.5 Disadvantages

There are two issues with Slope-Intercept form:

#### 1 Undefined Slope and Intercept

1. While using slope-intercept form, when $\Delta x = 0$, line will be $90^{\circ}$ to the x-axis.
2. This vertical line $90^0$ to x-axis has slope $m = \text{undefined}$ because $\tan (90^{\circ})$ is $\infty$ or undefined.
3. But a vertical line is a valid Classifier segregating data-points into two groups.
4. A Classifier cannot be defined using _undefined_ slope and _undefined_ intercept.
5. Hence an alternate equation for straight line called: **General Line Equation** is used instead of Slope-Intercept form.

Undefined slope and intercept: https://www.desmos.com/calculator/sbl976hwg0

#### 2 Limited to 2D

$$
\large
y = m \cdot x \;+\; c
$$

- Above equation only supports two variables $x$ and $y$
- Real world Machine Learning problems have large number of features.

## 4 General Line Equation (GLE)

### 4.1 Standard form

Equation of a straight line in **Standard form**.

$$
\large
a\,x \;+\; b\,y \;+\; c = 0
$$

GLE: https://www.desmos.com/calculator/odxm7i6azx

#### Properties

1. Generic form of a Linear equation.
2. GLE acts as a link between Coordinate Geometry and Linear Algebra for representing a line.

#### Standard form vs Slope-Intercept form

Transform Standard form into Slope-Intercept form:

Linear Equation of a straight line in Standard form:

$$
\large
w_1\,x \;+\; w_2\,y \;+\; w_0 = 0
$$

Rearranging above equation in slope-intercept form:

$$
\large
\begin{aligned}
w_1\,x \;+\; w_2\,y \;+\; w_0 &= 0 \\[8pt]
                       w_2\,y &= -\,w_1\,x \;\;+\;\; -\,w_0 \\[8pt]
                            y &= \biggl(-\,\frac{w_1}{w_2}\biggr)\,x \;+\; \biggl(-\,\frac{w_0}{w_2}\biggr)
\end{aligned}
$$

Yields Slope and Intercept formula as:

$$
\large
m = -\;\frac{w_1}{w_2} \quad\qquad c = -\;\frac{w_0}{w_2}
$$

### 4.2 Linear Expansion of GLE

#### Equation #1

General line equation for a line a.k.a **2D Hyperplane**:

$$
\large
w_1 x_1 + w_2 x_2 + w_0 = 0
$$

#### Equation #2

General line equation for a plane a.k.a **3D Hyperplane**:

$$
\large
w_1 x_1 + w_2 x_2 + w_3 x_3 + w_0 = 0
$$

#### Equation #3

General line equation for a Hyperplane a.k.a **$n$D Hyperplane**:

$$
\large
w_1 x_1 + w_2 x_2 + \ldots +  w_n x_n + w_0 = 0
$$

#### Exercise #1

### 4.3 Advantages of GLE

#### Why General Line Equation?

- A Linear Equation can be represented in either Slope-Intercept form or General Line Equation (GLE).
- GLE is preferred over Slope-Intercept form to overcome below two limitations in Slope-Intercept form:
  1. Undefined Slope
  2. Limited to 2D

#### 1 Perpendicular Line

- GLE simplifies the complexity around undefined slope.
- Undefined slope is solved by setting the coefficient of $y$ i.e., $b$ to zero in GLE: $a x + b y + c = 0$.
- Perpendicular line: https://www.desmos.com/calculator/iux1zvd92g

#### 2 Linear Expansion

- GLE supports scaling of Linear Equation into higher dimensions (beyond 2D Hyperplane) using Linear Expansion technique.
- For example refer Equation #2 (for 3D Hyperplane) and Equation #3 for (> 3D Hyperplane).
- 3D Hyperplane: https://www.desmos.com/3d/5o2v8quhbe

## 5 ML Context

### 5.1 Concepts

Concepts discussed above in Coordinate Geometry and Linear Algebra is revisited in ML Context.

#### Decision Boundary

Equation of line represents the Decision Boundary

$$
\large
w_1 x_1 + w_2 x_2 + w_0 = 0
$$

1. Decision Boundary := Line that segregates data-points (say on a scatter plot) into different classes/labels.
2. Binary classification has **one Decision Boundary**.
3. Multi-class classification has **many Decision Boundaries**.
4. In Binary classification Decision Boundary divides the Cartesian plane into two half spaces.

#### Variables

- $(w_0,\,w_1,\,w_2,\,\ldots,\,w_n)$ are called as **Weights** or parameters.
- $(x_0,\,x_1,\,x_2,\,\ldots,\,x_n)$ are **Feature** values.
- $w_0$ is a special type of weight called as **Bias**.

#### Weights

1. Weights ($w_0,\,w_1,\,w_2,\,\ldots,\,w_n$) are trainable parameters.
2. $w_0$ is called as **Bias**.
3. Before starting the Training process, weights are initialized with some random values.
4. Optimization process finds best weights for the line equation that segregates data-points based on their class labels.

#### Features

1. Features $(x_0, x_1, x_2, \ldots, x_n)$ are labeled data-points in the dataset.
2. Label column in the dataset acts as target variable $y$.

#### Learning

1. Learning in ML is the process of mathematically identifying best values for weights: $(w_0, w_1, w_2, \ldots, w_n)$ parameters.
2. Decision boundary resulting from this best values of weights combined with features is called as **best fit line**.

### 5.2 Binary Classification

#### How Binary Classification works?

How Binary Classifier uses below Decision Boundary in Binary Classification?

$$
\large
w_1 x_1 + w_2 x_2 + w_0 = 0
$$

Label value between $\{-1, +1\}$ is decided based on the value resulting from solving the decision boundary equation as shown below:

$$
\large
\begin{aligned}
y = f(x) =
\begin{cases} 
+1 \quad \text{if} \;\; w_1 x_1 + w_2 x_2 + w_0 > 0  \\[10pt]
-1 \quad \text{if} \;\; w_1 x_1 + w_2 x_2 + w_0 < 0
\end{cases}
\end{aligned}
$$

> **Note**:
>
> When $w_1 x_1 + w_2 x_2 + w_0 = 0$ it can be labeled as $-1$ or $+1$ based on the problem statement.

#### Example #1: Only two features

Mapping function for binary classification model having two features:

$$
\large
f(x) = w_1 x_1 + w_2 x_2 + w_0
$$

Decision Boundary for binary classification model having two features:

$$
\large
w_1 x_1 + w_2 x_2 + w_0 = 0
$$

#### Example #2: More than two features

Mapping function for binary classification model having $n$ features (where $n > 2$):

$$
\large
f(x) = w_1 x_1 + w_2 x_2 + \ldots +  w_n x_n + w_0
$$

Decision Boundary for binary classification model having $n$ features (where $n > 2$):

$$
\large
w_1 x_1 + w_2 x_2 + \ldots +  w_n x_n + w_0 = 0
$$

1. Features: $(x_0,\,x_1,\,x_2,\,\ldots,\,x_n)$ is given to us in the form of dataset.
2. An ML model learns Weights: $(w_0,\,w_1,\,w_2,\,\ldots,\,w_n)$ during learning process.

### 5.3 Half-spaces

#### Classification Context

In the context of a binary classification problem,

1. The decision boundary separates the data-points into two half-spaces.
2. One side is called as **Positive half-space**.
3. Other side is called as **Negative half-space**.

#### 1 Positive Half-space

On substituting the values of features in the Decision Boundary, if the value is greater than zero then the data-point lies in the positive half-space.

$$
\large
w_1 x_1 + w_2 x_2 + w_0 > 0
$$

#### 2 Negative Half-space

On substituting the values of features in the Decision Boundary, if the value is less than zero then the data-point lies in the negative half-space.

$$
\large
w_1 x_1 + w_2 x_2 + w_0 < 0
$$

#### On Decision Boundary

When substituting the values of features in the Decision Boundary, if the value is equal to zero then the data-point lies on the Decision Boundary.

$$
\large
w_1 x_1 + w_2 x_2 + w_0 = 0
$$
