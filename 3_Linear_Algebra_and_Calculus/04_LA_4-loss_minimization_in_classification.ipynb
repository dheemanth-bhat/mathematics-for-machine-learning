# Linear Algebra 4 - Loss minimization in classification

## 1 Decision Boundary

### 1.1 Half-space

#### Positive Half-space

In Positive Half-space all data-points have **distance greater than zero** because,

$$
\large
w_1 x_1 + w_2 x_2 + \ldots +  w_d x_d + w_0 > 0
$$

#### Negative Half-space

In Negative Half-space all data-points have **distance less than zero** because,

$$
\large
w_1 x_1 + w_2 x_2 + \ldots +  w_d x_d + w_0 < 0
$$

#### Decision Boundary

On Decision Boundary all data-points have **distance equal to zero** because,

$$
\large
w_1 x_1 + w_2 x_2 + \ldots +  w_d x_d + w_0 = 0
$$

### 1.2 Best Decision Boundary

There are two ways to identify the best line (Decision Boundary) that segregates the data-points into positive and negative half-spaces:

1. Using Direction/Angle
2. Using Distance/Magnitude

#### 1 Using Direction

1. Weight vector will always be perpendicular to the line (Decision Boundary).
2. All the data-points that lies in the opposite direction of the weight vector will be Negative i.e., $\cos \uptheta$ between (90, 270).

#### 2 Using Distance

When the distance is:

1. Greater than zero, data-points are in the Positive Half-space.
2. Less than zero, data-points are in the Negative Half-space.
3. Equal to Zero, data-points lies on the Decision Boundary and can be considered Positive or Negative class depending on the usecase.

### 1.3 Idea of Best-fit-line

#### What is a Best-fit-line?

#### How to identify a Best-fit-line?

- While identifying a best fit line, a Line can be considered as best Decision Boundary if the **average distance**  
  of all data-points from either side of the half-spaces is **maximum** compared to all other possible lines.
- Penalizing miss-classification.

#### Mathematical Notation

$$
\large
\begin{aligned}
g(x) &= \sum_{i=1}^n\,d_i \\[10pt]
g(x) &= \sum_{i=1}^n\,\frac{{w^T}{x^{(i)}}\;+\;w_0}{\Vert{w}\Vert}
\end{aligned}
$$

Where,

- $g(x)$ - Gain function of $x$
- $n$ - Number pf data-points in the dataset
- $d$ - distance of a data-point from the Decision Boundary
- $w$ - Weight vector
- $w_0$ - Bias
- $\displaystyle x^{(i)}$ - ith data-point, where $i$ ranges from 1 to $n$.

#### What is Gain function?

1. The above summation value is expected as, **larger the better**.
1. Distance in positive half-space is greater than zero.
1. Distance in negative half-space is less than zero.
1. But while calculating magnitude/distance absolute value is taken and their signs are neglected.
1. Hence the above summation value is decided as, larger the better.
1. Because we want to **maximize the gain** (in this case distance) its called **Gain function**.

#### Properties

1. Farther the distance of the data-points from the line on either side of the half-space, better the separator (Decision Boundary).
2. Positive and negative **distances cancel out each other** if they are directly summed up.
3. There are three tricks to resolve the problem of canceling distances.
   1. Absolute values of distances $\vert{d}\vert$
   2. Square the distances
   3. Multiply class label $y_i$ to each distance $d_i$
4. **Absolute** operation in a function $y = f(x) = \vert{x}\vert$ makes it **undifferentiable** hence Trick #1 is not used.
5. Even though **Square** operation is differentiable, it is **computationally expensive** compared to Trick #3.
6. Trick #3 is used in Optimization function since it correctly penalizes the misclassification compared to other two tricks.

### 1.4 Penalizing Misclassification

#### What is Misclassification?

when the actual label $y_i$ and the predicted label $\hat{y}$ are not same the data-point is miss-classified.

#### How to detect Misclassification?

#### How to penalize Misclassification?

$$
\large
g(x) = \sum_{i=1}^n\,\Biggl(\frac{{w^T}{x^{(i)}}\;+\;w_0}{\Vert{w}\Vert}\Biggr)\,\cdot\,y^{(i)}
$$

Where,

- $g(x)$ - Gain function of $x$
- $n$ - Number pf data-points in the dataset
- $w$ - Weight vector
- $w_0$ - Bias
- $x^{(i)}$ - ith row/data-point in dataset, where $i$ ranges from 1 to $n$
- $y^{(i)}$ - ith label in the labels column

Multiplying class label $y_i$ to each distance $d_i$ (Trick #3 from above listed techniques) simultaneously solves below two problems:

1. Resolves the canceling distances (sign of class label cancels the sign of the distance)
2. Penalizes misclassification.

##### Case #1: True Positive

- Class label is actually positive: $+1$
- Distance calculation resulted positive: $+d_i$
- Final result: $(+1) \times (+d_i) = +d_i$

##### Case #2: True Negative

- Class label is actually negative: $-1$
- Distance calculation resulted negative: $-d_i$
- Final result: $(-1) \times (-d_i) = +d_i$

##### Case #3: False Positive

- Class label is actually negative: $-1$
- Distance calculation resulted positive: $+d_i$
- Final result: $(-1) \times (+d_i) = -d_i$

##### Case #4: False Negative

- Class label is actually positive: $+1$
- Distance calculation resulted negative: $-d_i$
- Final result: $(+1) \times (-d_i) = -d_i$

#### Conclusion

1. Distances will be negative only for misclassification (False Positive and False Negative cases).
2. Distances will be positive only for correct classification (True Positive and True Negative cases).
3. Using this gain function we can **judge a better classifier** from bunch of classifier.

## 2 Mathematical Notations

Mathematical notations for a Binary Classification problem.

Problem Statement of a Binary Classification problem has two main components:

1. Dataset (Features and Labels)
2. An untrained Classifier

### 2.1 Dataset

#### Properties

- Dataset is made up of two components: Features $x$ and Class-labels $y$.
- Features can have any number of columns greater than one.
- Labels are mostly assigned to each data-point in one single column.
- Features $x$ and Class-labels $y$ are fixed components of the problem statement.

#### Mathematical Notation

$$
\large
D = \biggl\{\bigl(x_i\,,\;y_i\bigr)\biggr\}_{i=1}^n
$$

Where,

- $D$ is Dataset
- $x$ are features
- $y$ are labels
- $n$ is the size of the dataset

### 2.2 Labels

#### Properties

- Any type of boolean pairs can be used as class labels. For example: {True, False}, {0, 1}, {-1, +1}, etc.
- But {-1, +1} are chosen to represent Negative and Positive halfspaces respectively. 

#### Mathematical Notation

Actual Label $y$ can take values:

$$
\large
\begin{aligned}
y_i 
\begin{cases} 
+1 \quad \text{in +ve Half-Space} \\[15pt]
-1 \quad \text{in -ve Half-Space} 
\end{cases}
\end{aligned}
$$

### 2.3 Classifier

#### Mathematical Notation

Equation of Decision Boundary:

$$
\large
\begin{aligned}
w_1\,x_1\;+\;w_2\,x_2\;+\;\ldots\;+\;w_n\,x_n\;+\;w_0 &= 0 \\[3pt]
              \text{or} &    \\[3pt]
            w^T\,x\;+\;w_0 &= 0
\end{aligned}
$$

Predicted value $\hat{y}_i$ can take values:

$$
\large
\begin{aligned}
\hat{y}_i
\begin{cases} 
+1 \quad \text{if} \; \bigl(w^T\,x\;+\;w_0\bigr) \ge 0  \\[15pt]
-1 \quad \text{if} \; \bigl(w^T\,x\;+\;w_0\bigr) < 0 
\end{cases}
\end{aligned}
$$

### 2.4 Gain Function

#### Formula

Equation of Gain Function for Binary Classification:

$$
\large
g(D,\,\vec{w},\,w_0) = \frac{1}{n}\,\sum_{i=1}^n\,\Biggl(\frac{w^T\,x^{(i)}\;+\;w_0}{\Vert{w}\Vert}\Biggr)\,\cdot\,y^{(i)}
$$

#### Objective

- **Maximize Gain function**.
- Since $x^{(i)}$ and $y^{(i)}$ cannot be changed, identify best set of weights $w$ and bias $w_0$  
  (that represents different hyper-planes) that gives maximum value for the above equation.
- Maximizing Gain function is done via Optimization.

#### Maximization

Minimization is the process of optimizing the loss function: $l(D,\,\vec{w},\,w_0)$ to find the best weights: $w^*\,,\;w_0^*$ such that the loss function takes the smallest possible value.

$$
\large
\begin{aligned}
w^*\,,\;w_0^* &= \arg\max_{w,\,w_0}\;\;g(D,\,\vec{w},\,w_0) \\[5pt]
              &= \arg\max_{w,\,w_0}\;\frac{1}{n}\,\sum_{i=1}^n\,\Biggl(\frac{w^T\,x^{(i)}\;+\;w_0}{\Vert{w}\Vert}\Biggr)\,\cdot\,y^{(i)}
\end{aligned}
$$

Maximization of Gain function is the process of finding wights $w\,,\,w_0$ where the Gain function $g(D,\,\vec{w},\,w_0)$ takes largest possible value.

### 2.5 Loss Function

#### Definition

Negative Gain function is called as Loss function:

$$
\large
\text{Loss Function} = -\;\text{Gain Function}
$$

#### Formula

Equation of Loss Function for Binary Classification:

$$
\large
l(D, \vec{w}, w_0) = -\frac{1}{n}\,\sum_{i=1}^n\,\Biggl(\frac{{w^T}{x^{(i)}}\;+\;w_0}{\Vert{w}\Vert}\Biggr)\,\cdot\,y^{(i)}
$$

#### Objective

- **Minimize Loss function**.
- Identify best set of weights $w$ and bias $w_0$ that gives minimum value for the above equation.
- Minimizing Loss function is done via Optimization.

#### Minimization

Minimization is the process of optimizing the loss function: $l(D,\,\vec{w},\,w_0)$ to find the best weights: $w^*\,,\;w_0^*$ such that the loss function takes the smallest possible value.

$$
\large
\begin{aligned}
w^*\,,\;w_0^* &= \arg\min_{w,\,w_0}\;\;l(D,\,\vec{w},\,w_0) \\[5pt]
              &= \arg\min_{w,\,w_0}\;-\frac{1}{n}\,\sum_{i=1}^n\,\Biggl(\frac{w^T\,x^{(i)}\;+\;w_0}{\Vert{w}\Vert}\Biggr)\,\cdot\,y^{(i)}
\end{aligned}
$$

#### Properties

1. Loss function is also considered as **Error**.
2. Percentage of miss-classified observations can also be considered as Error.

## 3 Perceptron Learning

### 3.1 Handling Miss-Classification

#### What is Misclassification?

when the actual label $y_i$ and the predicted label $\hat{y}_i$ are not same the data-point is misclassified.

#### How to rectify Misclassification?

##### Rotate Clockwise

Update logic to rotate weight vector $\vec{w}$ in **clockwise direction**:

$$
\large
\vec{w}_{new} = \vec{w}_{old} \;+\; \vec{x}
$$

##### Rotate Anticlockwise

Update logic to rotate weight vector $\vec{w}$ in **anticlockwise direction**:

$$
\large
\vec{w}_{new} = \vec{w}_{old} \;+\; -\vec{x}
$$

#### How label helps in rectifying Misclassification?

1. If a **label is positive** but data-point is classified in **Negative half-space**, rotate the line in **clockwise** direction.
2. If a **label is negative** but data-point is classified in **Positive half-space**, rotate the line in **anticlockwise** direction.

### 3.2 Perceptron Learning Algorithm (PLA)

#### Weight Update Logic

Logic in Perceptron Learning Algorithm to update Weights for miss-classified data-point:

$$
\large
\begin{aligned}
&\text{if}\;\;\text{sign}\biggl(\frac{w^T x^{(i)} + w_0}{\Vert{w}\Vert} \biggr)\;\ne\;\text{sign}\bigl(y^{(i)}\bigr): \\[10pt]
&\qquad \quad w = w + x^{(i)} \cdot y^{(i)} \\[2pt]
&\qquad \quad w_0 = w_0 + y^{(i)}
\end{aligned}
$$
