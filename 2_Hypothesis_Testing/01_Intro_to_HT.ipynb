# Intro to Hypothesis Testing

## 1 Hypothesis Testing

### 1.1 What is Hypothesis Testing?

#### Definition

Hypothesis Testing is a branch of Inferential Statistics that helps us in breaking down the data and take trustable decisions.

#### Properties

1. Hypothesis Testing uses sample data as **evidence** to evaluate some **claims** made about population data and help us in taking **decisions** with some quantifiable **risk**.
2. Hypothesis Testing helps us in quantifying the **statistical significance** of values seen (as part of Descriptive Statistics) in the data.
3. In other words, for the given data, values that we see in Descriptive Statistics are statistically significant or not?
4. Hypothesis Testing is **not about predictions** instead it is trying to **infer** based on some evidence.

#### How Hypothesis Testing Works?

1. Make a **claim** against existing data by opposing some default assumption.
2. Perform Controlled **experiment**.
3. **Collect** data/evidence to support the claim.
4. **Evaluate** the significance of the evidence, backing the claim, against default assumption.
5. **Conclude** the Hypothesis Test.

##### Example

- Claim: "A new drug lowers blood pressure."
- Experiment: Randomized and **controlled** clinical trials on volunteers.
- Collect data: **Sampling** of Blood pressure measurements taken from volunteers.
- Analyze: Perform a statistical test.
- Conclusion: Determine whether the evidence supports the claim.

#### What are the outcomes of Hypothesis Testing?

- When there is **enough data or evidence** to backup the claim then the **default assumption is rejected**.
- When the **data or evidence** backing the claim is **insignificant** we **fail to reject the default assumption**.

> **Note**:
> 
> Burden of proof is on those who want to reject default assumption.

#### Hypothesis testing Framework

Hypothesis -> Experiment (controlled) -> Handful (Sampling) -> Unbiased -> Experiment output -> Validation

Hypothesis Testing Framework is the process of designing some experiments, collecting data as evidence to **measure the impact** of the experiment.

##### Types of Impact

There can be three types of impact:

1. No Impact
2. Positive Impact
3. Negative Impact

Hypothesis testing tries to validate the **significance** of claimed positive or negative impact over default current state of no impact.

### 1.3 Null vs Alternate Hypothesis

#### Definition

1. Default assumption is called as Null Hypothesis and it is denoted as: $H_0$
2. Opposite of Default assumption is called as Alternate Hypothesis and is denoted as: $H_a$

#### Examples

##### Examples for Null Hypothesis

1. Coin is fair (not biased).
2. Suspect is innocent (not guilty).
3. Tumor is benign hence patient has no cancer.

##### Examples for Alternate Hypothesis

1. Coin is t biased.
2. Suspect is guilty.
3. Tumor is malignant hence patient has cancer.

#### How to define Null Hypothesis?

- Null hypothesis should be defined such that it is **possible to calculate the probability of observing data** given $H_0$ is true.
- For example in a coin toss experiment, we can take probability of observing 3 consecutive heads in 5 toss, when we formulate $H_0$  
   as "coin is fair" as oppose to "coin is unfair", because it is not possible to calculate probability when the coin is unfair to start with.

#### How Assumptions are used in HT Framework?

Assumptions formulated in the beginning play crucial role in decision-making step of Hypothesis Testing Framework:

- First step in Hypothesis Testing Framework is to define the two types of Assumptions:
  - Default Assumption $H_0$
  - Alternate Assumption $H_a$
- End of Hypothesis Testing will have either one of the below two outcomes:
  1. Successfully reject $H_0$ or accept $H_a$
  2. Successfully reject $H_a$ or failed to reject $H_0$

#### Properties

1. There is no term or process to **accept Null Hypothesis**. Null Hypothesis is **de facto** and is already accepted.
2. Gather evidence towards the claim to either **reject Null Hypothesis** or **failed to reject Null Hypothesis**.
3. Gather evidence towards the claim to either **accept Alternate Hypothesis** or **reject Alternate Hypothesis**.

### 1.4 p-value

#### Definition

$\text{p-value}$ is **probability of observing data** as extreme as what was observed given that the Null Hypothesis **$H_0$ is true**.

#### Equation

Mathematical representation of $\text{p-value}$:  

$$
\large
\begin{aligned}
\text{p-value} = P(\text{data} \mid H_0\;\text{is true})
\end{aligned}
$$

#### Properties

- $\text{p-value}$ is the metric used in the decision-making step of Hypothesis Testing Framework.
- $\text{p-value}$ is a probability value ranging between 0 and 1.
- $\text{p-value}$ is probability of all events which are more extreme than the given event that we are trying to analyze.
- $\text{p-value}$ is not probability of $H_0$ or $H_a$ i.e., $\text{p-value}$ does not directly measure the probability that $H_0$ is True.

#### How p-value is used in HT Framework?

- Since $\text{p-value}$ emphasizes on $H_0$ being True to start with, lower the $\text{p-value}$ lower will be the chances of $H_0$ being true.
- Smaller the $\text{p-value}$ higher is the confidence in our claim (Alternate Hypothesis $H_a$).
- We reject $H_0$ when $\text{p-value}$ is very low.
- Threshold for "very low" depends on the problem statement.

$$
\large
\begin{aligned}
\text{p-value}\,\downarrow&\quad\implies\quad\text{Evidence supporting}\,H_0\,\downarrow \\[5pt]
                          &\qquad\text{or}                                               \\[5pt]
\text{p-value}\,\downarrow&\quad\implies\quad\text{Evidence supporting}\,H_a\,\uparrow
\end{aligned}
$$

#### Example

Say we are convicting a suspect for murder, we start by formulating the Null Hypothesis $H_0$.

##### **STEP #1**: Formulate Null and Alternate Hypothesis

- $H_0$ : Suspect is innocent (not guilty)
- $H_a$ : Suspect is guilty of murder

##### **STEP #2**: Collect Evidences

- **Evidence #1**: Suspect has a knife in his pocket
- **Evidence #2**: Knife has blood stains
- **Evidence #3**: Blood stains matches with the victim
- **Evidence #4**: Fingerprint matches with the suspect
- **Evidence #5**: CCTV caught the suspect killing the victim

##### **STEP #3**: Calculate p-value

Metric $\text{p-value}$ is the sum of probability of all the evidences.

##### **STEP #4**: Decision

- **Probability** of seeing so many evidences given that the suspect is innocent will be **very low**.
- With every evidence probability of $H_0$ being true (suspect being innocent) reduces, resulting in a very low $\text{p-value}$.
- Since the $\text{p-value}$ is very low reject the Null Hypothesis $H_0$.
- Hence accept the Alternate Hypothesis $H_a$ that the suspect is guilty of murder.

### 1.5 Significance level and Confidence level

#### Definition

Significance level ($\alpha$) is the threshold value for $\text{p-value}$ used during the decision-making step of the Hypothesis Testing Framework.

#### Formula

Relationship between Significance level and Confidence level:

$$
\large
\begin{aligned}
\text{Significance level} &= 1 - \text{Confidence level} \\[5pt]
                          & \text{or} \\[5pt]
 \text{Confidence level}  &= 1 - \text{Significance level}
\end{aligned}
$$

#### Properties

1. Significance level is decided based on Confidence level.
2. Confidence level indicates how confident are we on the gathered evidence/data.
3. Value for Significance level can range between 0 and 1.

#### How Significance Level is used in HT Framework?

- Significance level or threshold value is used in the decision-making step of HT Framework.
- Significance level is part of if-else condition in the decision-making step of HT Framework.

$$
\large
\begin{aligned}
&\text{if p-value} \le \alpha \text{:}                       \\[3pt]
&\quad\quad \text{Reject $H_0$ i.e., Accept $H_a$}           \\[6pt]
&\text{else:}                                                \\[3pt]
&\quad\quad \text{Failed to reject $H_0$ i.e., Reject $H_a$}
\end{aligned}
$$

Where,

- $\alpha$ is called as **Significance level**
- $1 - \alpha$ is called as **Confidence level**

#### How to decide the threshold value?

1. Threshold value is selected in such a way that **chances of error** is less.
2. Threshold value $\alpha$ depends on the **domain** of the problem.
3. For example medical domain can have very low threshold of 0.01 or 1%.
4. Generally the value of $\alpha$ must be less than 0.05 or 5%.

> **Note**:
>
> Generally in most of the industries 95% confidence level is considered hence most  
> common value for Confidence Level is 0.95, therefore Significance Level is 0.05.

#### Risks involved in Significance level

- Setting a Significance level while taking a decision often involves some risk (or results in errors).
- There can be errors when calculating the $\text{p-value}$, which affects in the decision making.

Risk can be illustrated using a confusion matrix:

```
                      Actual
                   0         1     
             +----------+----------+
           0 | No Error |  Error   |
Predicted    +----------+----------+
           1 |  Error   | No Error |
             +----------+----------+
```

In the above diagram,

1. Its declared _No Error_ when the Actual and the Predicted values **are same**.
2. Its declared _Error_ when the Actual and the Predicted values **are different**.

#### How Confusion Matrix is used in HT Framework?

- Confusion Matrix just captures the errors and does not reduce the errors.

As seen in the above confusion matrix there are two types of errors:

1. False Positive in Confusion Matrix is mapped to **Type 1 Error** $(\alpha)$ a.k.a Significance level.
2. False Negative in Confusion Matrix is mapped to **Type 2 Error** $(\beta)$

### 2.3 Power of test

In a Machine learning system the goal is to **reduce both Type-1 and Type-2 errors**.

#### Definition

1. Power represents the probability of correctly rejecting a False Null Hypothesis.
2. $1 - \beta$ (True Positive) is also called as **Power of test**.

```
                        Actual
                    0           1         
             +------------+------------+
           0 | 1 - α (TN) |     β (FN) |
Predicted    +------------+------------+
           1 |     α (FP) | 1 - β (TP) |
             +------------+------------+
```

The goal of power of test is to reduce Type-2 error $(\beta)$.


$\Rightarrow$ $\text{FN} + \quad\;\;\text{TP} = 1$  
$\Rightarrow$ $\;\;\;\beta + (1 - \beta) = 1$


1. To reduce Type-2 error $\beta$ (False Negative) we have to increase $1 - \beta$ (True Positive).
2. To reduce Type-1 error $\alpha$ (False Positive) we have to increase $1 - \alpha$ (True Negative).

> **Note**:
>
> To increase Power of test $(1 - \beta)$ we have to reduce $\beta$.

#### Relationship between Alpha and Beta

$
\begin{align}
\large
\alpha = \frac{1}{\beta}
\end{align}
$

Since $\alpha$ and $\beta$ are inversely proportional to each other, mathematically it is not possible to reduce both the errors at the same time.  
We always need to find a trade off  between $\alpha$ and $\beta$.

#### Factors influencing Power

1. Sample size $(n)$: As sample size increases power increases.
2. Significance level $(\alpha)$: As Significance level increases beta reduces (hence power increases).
3. Variance $(\sigma)$: As Variance increases Type-2 error increases (hence power reduces).
4. Effect Size $(d)$: As effect size increases power increases.

#### Summary

<div style="display: inline-block">

|   # | Metric         |       Symbol | Representation     | Error Type   | Interpretation in Hypothesis Testing                        |
| --: | :------------- | -----------: | :----------------- | :----------- | :---------------------------------------------------------- |
|   1 | True Negative  | $1 - \alpha$ | Confidence level   |              |                                                             |
|   2 | False Negative |      $\beta$ |                    | Type-2 Error | Failing to reject an incorrect (False) Null Hypothesis.     |
|   3 | True Positive  |  $1 - \beta$ | Power of Test      |              | Probability of correctly rejecting a False Null Hypothesis. |
|   4 | False Positive |     $\alpha$ | Significance level | Type-1 Error | Incorrect rejection of a True Null Hypothesis.              |

</div>

## 3 Types of Tailed Test

Types of tests:

1. One Tailed Test
    1. Left Tailed Test
    2. Right Tailed test
2. Two Tailed test

### 3.1 Cases

#### Case #1

$H_0: \mu \ge value$  
$H_a: \mu \lt value$  

Perform Left-tailed test.

#### Case #2

$H_0 : \mu \le value$  
$H_a : \mu \gt value$  

Perform Right-tailed test.

#### Case #3

$H_0 : \mu = value$  
$H_a : \mu \ne value$  

Perform Two-tailed test.

> **Note**:
>
> Equal comparison i.e., $=$ operator should always be present in $H_0$

1. if condition is $<$ or $\le$ in $H_a$ then perform **Left-Tailed** test.
2. if condition is $>$ or $\ge$ in $H_a$ then perform **Right-Tailed** test.
3. if condition is $\ne$ in $H_a$ then perform **Two-Tailed** test.

#### Summary

<div style="display: inline-block">

| Case         | Tail test type    |
| :----------: | :---------------- |
| $<$ or $\le$ | Left tailed test  |
| $>$ or $\ge$ | Right tailed test |
| $\ne$        | Two tailed test   |

</div>

## 4 Hypothesis Testing Framework

### 4.1 Steps in Hypothesis Testing

- **STEP #1**: Formulate Null-Hypothesis $H_0$ and Alternate-Hypothesis $H_a$
- **STEP #2**: Choose a Distribution, Test-statistic and Significance-level $(\alpha)$
- **STEP #3**: Select type of test: Left / Right / Two-tailed
- **STEP #4**: Compute $\text{p-value}$
- **STEP #5**: Compare $\text{p-value}$ with $\alpha$, reject $H_0$ if $\text{p-value} \le \alpha$ else reject $H_a$

### 4.2 Examples

#### Coin Toss

##### STEP 1: Formulate Null and Alternate Hypothesis

- $H_0$ : Coin is fair
- $H_a$ : Coin is biased

Above formulation can be mathematically written as,

- $H_0 : p = 0.5$
- $H_a : p \ne 0.5$

##### STEP 2: Choose Significance level

````python
# Distribution: Binomial distribution
# Significance level: 0.05
alpha = 0.05
````

##### STEP 3: Select type of test

Since $H_a$ has $\ne$ symbol perform two tailed test.

##### STEP 4: Compute p-value

````python
p_value = 0.001  # Dummy value for example
````

##### STEP 5: Compare p-value with alpha

````python
if p_value < alpha:
    print("Reject Null-Hypothesis i.e., accept Alternate-Hypothesis.")
else:
    print("Failed to reject Null-Hypothesis i.e., reject Alternate-Hypothesis.")
````
