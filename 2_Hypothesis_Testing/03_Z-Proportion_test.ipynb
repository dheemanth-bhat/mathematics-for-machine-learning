# Z-Proportion test

````python
import numpy as np
import scipy as sp
import statsmodels.api as sm
````

## 1 Introduction to Z-Proportion Test

### 1.1 Assumptions for Z-Proportion test

1. Z-Proportion test is used for testing hypothesis on population parameter type - Proportion.
2. Sampling distribution of sample proportion is normal and its mean is available.

### 1.2 Conditions for Z-Proportion test

1. Random Sampling.
2. Observations are independent of each other.
3. Sample size must be large enough so the sampling distribution of the proportion is approximately normal.
   - $np_0 \ge 10$
   - $n(1 - p_0) \ge 10$
5. Known population standard deviation.

> **Note**:
>
> 1. For Z-Proportion test, prior knowledge of population standard deviation is not required.
> 2. Population standard deviation can be calculated using proportion.

### 1.3 Types of Z-Proportion tests

1. One Sample Z-Proportion test
2. Two sample Z-Proportion test

> **Note**:
>
> In Two sample Z-Proportion test the two samples can be of different size but must be i.i.d.  
> i.id := Independent and Identically Distributed

## 2 One Sample Z-Proportion test

### 2.1 Introduction

#### What is One Sample Z-Proportion test?

In One Sample Z-Proportion test, proportion of one single sample is compared against population proportion.

#### When to use One Sample Z-Proportion test?

- When the distribution of data is **Binomial Distribution**.
- When we have values of population parameters like proportions.
- When we want to compare proportion values between sample distribution and population distribution.

> **Note**:
>
> Z-Proportion test is used only on Binomial Distribution.

### 2.2 Nature of hypothesis

- $H_0: p = p_0$
- $H_a: p \neq p_0 \;(\,\text{or}\; p > p_0 \,,\, p < p_0\,)$

Where,

- $p$ - Variable representing Population Proportion
- $p_0$ - Value of Population Proportion

#### Examples

##### Example #1

- $H_0: p = 80\%$
- $H_a: p \neq 80\%$

##### Example #2

- $H_0: p = 80\%$
- $H_a: p < 80\%$

##### Example #3

- $H_0: p = 80\%$
- $H_a: p > 80\%$

### 2.3 Test Statistic

1. The name of the test statistic in One sample Z-proportion test is called as **Z-Statistic**.
2. Sample distribution of sample proportion follows Normal Distribution.

#### Reference Distribution

Reference Distribution is **Z Distribution** (Standard Normal Distribution) with mean $\mu = 0$ and standard deviation $\sigma = 1$.

#### Formula

Given the observed number of "success" $x$ and in sample size $n$, proportion $\widehat{p}$ can be calculated as:

$$
\large
\widehat{p} = \frac{x}{n}
$$

Formula to calculate $\text{Z-Statistic}$ for One Sample Z-Proportion test:

$$
\large
\text{Z-Statistic} = \frac{\widehat{p} - p_0}{\sqrt{\displaystyle  \frac{p_0\;(1\;-\;p_0)}{n}}}
$$

Where,

- $\widehat{p}$ - Observed value of Proportion (Sample Proportion)
- $p_0$ - Value of Population Proportion

> **Note**:
>
> $\text{Z-Statistic}$ is not same as $\text{Z-Score}$.

### 2.4 API

```python
import statsmodels.api as sm


# Compute Z-Statistic and p-value.
z_stat, p_value = sm.stats.proportions_ztest(
    count=x,  # Observed "success" count
    nobs=n,  # Sample size
    value=p,  # Population proportion
    alternative="two-sided",  # Use "less" for Left Tailed test and "greater" for Right Tailed test.
    prop_var=p,
)
```

## 3 Two Sample Z-Proportion test

### 3.1 Introduction

#### What is Two Sample Z-Proportion test?

In Two Sample Z-Proportion test, proportion of one sample is compared against proportion of another sample.

#### When to use Two Sample Z-Proportion test?

- When the distribution of sample data is **Binomial Distribution**.
- When we have values of population parameters like proportions.
- When we have to compare proportion values between two sample distributions from same or different population.

> **Note**:
>
> Z-Proportion test is used only on Binomial Distribution.

### 3.2 Nature of hypothesis

- $H_0: p_1 = p_2$
- $H_a: p_1 \neq p_2 \;(\,\text{or}\; p_1 > p_2 \,,\, p_1 < p_2\,)$

or,

- $H_0: p_1 - p_2 = 0$
- $H_a: p_1 - p_2 \neq 0 \;(\,\text{or}\; p_1 - p_2 > 0 \,,\, p_1 - p_2 < 0\,)$

Where,

- $p_1$ - Proportion value of Sample #1.
- $p_2$ - Proportion value of Sample #2.

### 3.3 Test Statistic

1. The name of the test statistic in Two sample Z-proportion test is called as Z-statistic.
2. Sample distribution of sample proportion follows Normal Distribution.

#### Reference Distribution

Reference Distribution is **Z Distribution** (Standard Normal Distribution) with mean $\mu = 0$ and standard deviation $\sigma = 1$.

#### Formula

For two samples $s_1$ and $s_2$, given the observed number of "success" $x_1$ and $x_2$ in sample sizes $n_1$ and $n_2$ respectively,  
Pooled proportion $\widehat{p}$ can be calculated as:

$$
\large
\widehat{p} = \frac{x_1 + x_2}{n_1 + n_2}
$$

Formula to calculate $\text{Z-Statistic}$ for Two Sample Z-Proportion test:

$$
\large
\text{Z-Statistic} = \frac{\widehat{p}_1 - \widehat{p}_2}{\sqrt{\widehat{p}\;\bigl(1-\widehat{p}\,\bigr)\;\Bigl(\frac{1}{n_1}+\frac{1}{n_2}\Bigr)}}
$$

Where,

- $p_1$ - Proportion value of Sample #1.
- $p_2$ - Proportion value of Sample #2.
- $\widehat{p}$ - Pooled Proportion.

### 3.4 API

```python
import statsmodels.api as sm


# Given,
tot_obs: list = []  # Total number of observations.
suc_count: list = []  # Number of "success" in total observations.

# Compute Z-Statistic and p-value.
z_stat, p_value = sm.stats.proportions_ztest(
    count=suc_count,
    nobs=tot_obs,
    alternative="two-sided",  # Use "less" for Left Tailed test and "greater" for Right Tailed test.
)
```

## 4 Quizzes

### Quiz #1

A fast-food restaurant claims that 80% of their customers prefer their new burger over the old one.  
In a random sample of 100 customers, 85 said they preferred the new burger.  

What is the null and alternative hypothesis?

A. Null Hypothesis $H_0: \text{p = 0.80}$, Alternate Hypothesis $H_1: \text{p $\neq$ 0.80}$  
B. Null Hypothesis $H_0: \text{p = 0.80}$, Alternate Hypothesis $H_1: \text{p < 0.80}$  
C. Null Hypothesis $H_0: \text{p = 0.85}$, Alternate Hypothesis $H_1: \text{p $\neq$ 0.85}$  
D. Null Hypothesis $H_0: \text{p = 0.85}$, Alternate Hypothesis $H_1: \text{p > 0.85}$  

#### Solution

Option A:  

A. Null Hypothesis $H_0: \text{p = 0.80}$, Alternate Hypothesis $H_1: \text{p $\neq$ 0.80}$ 

##### Hypothesis Testing

````python
# STEP #1:
# H_0: p = 80%  # 80% of customers prefer new burger
# H_a: p != 80%  # 80% of customers does not prefer new burger
````

````python
# STEP #2:
# Distribution: Standard Normal Distribution
# Test Statistic: Z-Statistic
# Significance level: 0.05
alpha = 0.05
````

````python
# STEP #3:
# Since H_a contains != symbol, perform two-tailed test.
````

````python
# Given,
p_0 = 0.8
n = 100  # Sample size
x = 85  # observed value

# Find: P(X != 0.8)
````

````python
p_hat = x / n
round(p_hat, 6)
````

````python
# STEP #4:
# Calculate test statistic.
z_stat = (p_hat - p_0) / np.sqrt((p_0 * (1 - p_0)) / n)
z_stat.round(6).item()
````

````python
sp.stats.norm.cdf(-np.abs(z_stat)).round(2).item()
````

````python
sp.stats.norm.cdf(z_stat).round(2).item()
````

````python
# Compute p-value.
p_value = 2 * (1 - sp.stats.norm.cdf(z_stat))  # 2-tailed test
p_value.round(6).item()
````

````python
# STEP #5:
# Compare p-value with significance level.
if p_value <= alpha:
    print("Reject Null Hypothesis.")
    print("80% of customers does not prefer new burger.")
else:
    print("Failed to reject Null Hypothesis.")
    print("80% of customers prefer new burger.")
````

### Quiz #2

You are a product manager for a company that has recently launched a new product.  

Customer satisfaction is a critical metric, and you want to determine if the proportion of  
satisfied customers with the new product meets your target satisfaction level of 70%.

You collected a random sample of 150 customer reviews, and 115 of them expressed satisfaction with the product.

#### Solution #1

Manual Calculation

````python
# STEP #1:
# H_0: p = 70%  # Proportion of satisfied customers meets the target of 70%
# H_a: p != 70%  # Proportion of satisfied customers does not meet the target of 70%
````

````python
# STEP #2:
# Distribution: Standard Normal Distribution
# Test Statistic: Z-Statistic
# Significance level: 0.05
alpha = 0.05
````

````python
# STEP #3:
# Since H_a contains != symbol, perform tow-tailed test.
````

````python
# Given,
p_0 = 0.7  # 70%
n = 150  # Sample size
x = 115  # Observed value

p_hat = x / n  # Observed Proportion
round(p_hat, 2)
````

Hypothesis Testing is done to quantify the significance of the observed proportion 0.77 (i.e., 77%) of satisfied customers.

````python
# STEP #4:
# Calculate test statistic.
z_stat = (p_hat - p_0) / np.sqrt((p_0 * (1 - p_0)) / n)
z_stat.round(6).item()
````

````python
# Compute p-value.
p_value = 2 * sp.stats.norm.cdf(-np.abs(z_stat))
p_value.round(6).item()
````

````python
# STEP #5:
# Compare p-value with significance level.
if p_value <= alpha:
    print("Reject Null hypothesis.")
    print("Proportion of satisfied customers does not meet the target of 70%.")
else:
    print("Failed to reject Null hypothesis.")
    print("Proportion of satisfied customers meets the target of 70%.")
````

#### Solution #2: Using library

````python
# STEP #1:
# H_0: p = 70%  # Proportion of satisfied customers meets the target of 70%
# H_a: p != 70%  # Proportion of satisfied customers does not meet the target of 70%

# STEP #2:
# Distribution: Standard Normal Distribution
# Test Statistic: Z-Statistic
# Significance level: 0.05
alpha = 0.05  # Default value

# STEP #3:
# Since H_a contains != symbol, perform two-tailed test.

# Given,
p = 0.7  # Population proportion
n = 150  # Sample size
x = 115  # Observed proportion

# STEP #4:
# Compute test statistic and p-value.
z_stat, p_value = sm.stats.proportions_ztest(
    count=x,
    nobs=n,
    value=p,
    alternative="two-sided",  # Pass "less" for Left-tailed-test, "greater" for Right-tailed-test.
    prop_var=p,
)
print("z-statistic:", z_stat.round(6).item())
print("p-value:", p_value.round(6).item())

# STEP #5:
# Compare p-value with alpha.
if p_value <= alpha:
    print("Reject Null hypothesis.")
    print("Proportion of satisfied customers does not meet the target of 70%.")
else:
    print("Failed to reject Null hypothesis.")
    print("Proportion of satisfied customers meets the target of 70%.")
````

### Quiz #3

You are the manager of an e-commerce website, and you have recently implemented a new web page in hopes of increasing sales.  
To evaluate the effectiveness of the new page, you collected data on the conversion rates for both the old and new web pages. The conversion rate is defined as the proportion of visitors who make a purchase.

- For the old web page (Web Page A), you had **1000** visitors, resulting in **50** conversions.
- For the new web page (Web Page B), you had **500** visitors, resulting in **30** conversions.

Now, you want to determine if there is a statistically significant difference in the conversion rates between the old and new web pages.

#### Solution #1: Using Library

````python
# STEP #1:
# H_0: p_1 = p_2  # No change in conversion rate.
# H_a: p_1 != p_2  # This is some significant change in conversion rate.
````

````python
# STEP #2:
# Distribution: Standard Normal Distribution
# Test Statistic: Z-Statistic
# Significance level: 0.05
alpha = 0.05
````

````python
# STEP #3:
# Since H_a contains != symbol perform two tailed test.
````

````python
# Given,
# Number of visits for Web Page A and Web Page B # n_1, n_2
obs_vals = np.array([1000, 500])

# Number of conversions for Web Page A and Web Page B # x_1, x_2
smpl_sizes = np.array([50, 30])
````

````python
# STEP #4:
# Calculate test-statistic and p-value.
z_stat, p_value = sm.stats.proportions_ztest(
    count=smpl_sizes,
    nobs=obs_vals,
    alternative="two-sided",
)
print("Z-Statistic:", z_stat.round(6).item())
print("p-value", p_value.round(6).item())
````

````python
# STEP #5:
# Compare p-value with alpha.
if p_value <= alpha:
    print("Reject Null Hypothesis. Change in conversion rate is significant.")
else:
    print("Failed to reject Null Hypothesis. No change in conversion rate.")
````

#### Solution #2

````python
# STEP #1:
# H_0: p_1 = p_2  # No change in conversion rate.
# H_a: p_1 != p_2  # This is some significant change in conversion rate.
````

````python
# STEP #2:
alpha = 0.05  # Significance level
````

````python
# STEP #3:
# Since H_a contains != symbol perform two tailed test.
````

````python
# Given,
x_1 = 50
n_1 = 1000
x_2 = 30
n_2 = 500
````

````python
p_1 = x_1 / n_1
round(p_1, 6)
````

````python
p_2 = x_2 / n_2
round(p_2, 6)
````

````python
p_hat = (x_1 + x_2) / (n_1 + n_2)
round(p_hat, 6)
````

````python
# Calculate pooled proportion.
pooled_p = p_hat * (1 - p_hat) * ((1 / n_1) + (1 / n_2))

# Standard error using pooled proportion.
se_pooled = np.sqrt(pooled_p)
se_pooled.round(6).item()
````

````python
# NOTE: (Optional) Not used in below Z-Statistic calculation.

# Calculate un-pooled proportion.
unpold_p = (p_1 * (1 - p_1) / n_1) + (p_2 * (1 - p_2) / n_2)

# Standard error using un-pooled proportion.
se_unpold = np.sqrt(unpold_p)
se_unpold.round(6).item()
````

````python
# STEP 4:
# Compute Z-statistic and p-value.
z_stat = (p_1 - p_2) / se_pooled
z_stat.round(6).item()
````

````python
p_value = 2 * sp.stats.norm.cdf(-np.abs(z_stat))
p_value.round(6).item()
````

````python
# STEP #5:
# Compare p-value with alpha.
if p_value <= alpha:
    print("Reject Null Hypothesis. Change in conversion rate is significant.")
else:
    print("Failed to reject Null Hypothesis. No change in conversion rate.")
````
